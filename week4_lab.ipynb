```python
number={1,2,3,4,3,3}
number.add(5)
number.remove(3)
number.discard(3)#no error if element not found
print("set:",number) 
```

```python
#set operations
number={1,2,3,4,3,3}
evens={2,4,6,8}
print("intersection:",number & evens)
print("updated set:",number)

```
