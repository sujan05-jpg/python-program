```python
fruits=["apple","banana","cherry","apple"]

print("original list:",fruits)

fruits.append("orange")#add elements

print("updated list",fruits)

if"banana" in fruits:
    fruits.remove("banana")#removes the fruits
    print("updated list",fruits)
    #access by the indexing
print("first fruit:",fruits[2])
## list comprehension
upper_fruits=[fruit.upper() for fruit in fruits]
print("uppercase fruits:",upper_fruits)
```
